# Dice simulations

Simulations of one die,

    a) Calculate theoretical mean of a dice (six-sided)

    b) Now do simulations of:

- 10
- 100
- 1000
- 10000
- 100000
- 1000000
- 10000000

number of dices and plot their means.

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# a) Calculate theoretical mean of a dice (six-sided)

dice_values = [1, 2, 3, 4, 5, 6]
probability = 1/6

expected_values = [value * probability for value in dice_values]

theoretical_mean = sum(expected_values)
print(theoretical_mean)
```

```python
simulations = [10, 100, 1_000, 10_000, 100_000, 1_000_000]

def dice_roll():
    return random.randint(1, 6)

def calculate_mean(num_of_simulations):
    rolls = []
    for i in range(num_of_simulations):
        rolls.append(dice_roll())
    return np.mean(rolls)

# Lists to store simulation counts and their corresponding means
simulation_counts = []
means = []

for simulation in simulations:
    mean = calculate_mean(simulation)  # Calculate mean only once per loop
    simulation_counts.append(simulation)
    means.append(mean)
    print(mean)

plt.scatter(simulations, means, s=50, zorder=3)
plt.xscale('log')
plt.ylim([0, 6])
plt.xlabel('Number of Simulations (Log Scale)')
plt.ylabel('Mean Value')
plt.title('Mean Dice Value vs. Number of Simulations')
plt.grid()
plt.show()
```
