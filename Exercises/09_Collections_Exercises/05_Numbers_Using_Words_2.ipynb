# Print a given number using words (version 2)

Make the program ask the user to input any number from 0 to 99999.

The program should then print out the full sequence of words for the number.

Example:

- Input: 34507
- Output: Thirtyfour thousand five hundred and seven

```python
numbers_in_words = {
    "0": "Zero",
    "1": "One",
    "2": "Two",
    "3": "Three",
    "4": "Four",
    "5": "Five",
    "6": "Six",
    "7": "Seven",
    "8": "Eight",
    "9": "Nine",
    "10": "Ten",
    "11": "Eleven",
    "12": "Twelve",
    "13": "Thirteen",
    "14": "Fourteen",
    "15": "Fifteen",
    "16": "Sixteen",
    "17": "Seventeen",
    "18": "Eighteen",
    "19": "Nineteen"
}

tens_in_words = {
    "2": "Twenty",
    "3": "Thirty",
    "4": "Forty",
    "5": "Fifty",
    "6": "Sixty",
    "7": "Seventy",
    "8": "Eighty",
    "9": "Ninety"
}

thousands = "Thousand"
hundred = "Hundred"
and_word = "and"

def convert_to_words(num_str):
    if 10 <= int(num_str) <= 19:
        return numbers_in_words[num_str]
    elif num_str[0] != "0":
        return tens_in_words[num_str[0]] + numbers_in_words[num_str[1]].lower()
    else:
        return numbers_in_words[num_str[1]]

user_input = input("Please enter a number from 0 to 99999:")

if 0 <= int(user_input) <= 99999:
    user_input_words = ""
    
    # Thousands
    if len(user_input) >= 5:
        user_input_words += convert_to_words(user_input[-5:-3]) + " " + thousands + " "
    elif len(user_input) == 4 and user_input[-4] != "0":
        user_input_words += numbers_in_words[user_input[-4]] + " " + thousands + " "
    
    # Hundreds
    if len(user_input) >= 3 and user_input[-3] != "0":
        user_input_words += numbers_in_words[user_input[-3]] + " " + hundred + " "
    
    # Check if 'and' is needed
    if len(user_input) > 2 and user_input[-2:] != "00":
        user_input_words += and_word + " "
    
    # Tens & Ones
    if len(user_input) >= 2:
        user_input_words += convert_to_words(user_input[-2:])
    elif len(user_input) == 1:
        user_input_words += numbers_in_words[user_input[-1]]
    
    result = user_input_words[0].upper() + user_input_words[1:].lower()
    
    print(f'{user_input = }')
    print(result.strip())
else:
    print("Please enter a number between 0 and 99999.")
```

```python
numbers_in_words = {
    "0": "Zero",
    "1": "One",
    "2": "Two",
    "3": "Three",
    "4": "Four",
    "5": "Five",
    "6": "Six",
    "7": "Seven",
    "8": "Eight",
    "9": "Nine",
    "10": "Ten",
    "11": "Eleven",
    "12": "Twelve",
    "13": "Thirteen",
    "14": "Fourteen",
    "15": "Fifteen",
    "16": "Sixteen",
    "17": "Seventeen",
    "18": "Eighteen",
    "19": "Nineteen"
}

tens_in_words = {
    "2": "Twenty",
    "3": "Thirty",
    "4": "Forty",
    "5": "Fifty",
    "6": "Sixty",
    "7": "Seventy",
    "8": "Eighty",
    "9": "Ninety"
}

thousands = "Thousand"
hundred = "Hundred"
and_word = "and"

def convert_to_words(num_str):
    if 10 <= int(num_str) <= 19:
        return numbers_in_words[num_str]
    elif num_str[0] != "0":
        return tens_in_words[num_str[0]] + numbers_in_words[num_str[1]].lower()
    else:
        return numbers_in_words[num_str[1]]

def convert_thousands(num):
    if len(num) == 5:
        return convert_to_words(num[0:2]) + " " + thousands + " "
    elif len(num) == 4:
        return numbers_in_words[num[0]] + " " + thousands + " "
    return ""

def convert_hundreds(num):
    if len(num) >= 3 and num[-3] != "0":
        return numbers_in_words[num[-3]] + " " + hundred + " "
    return ""

def convert_tens_ones(num):
    return convert_to_words(num[-2:]) if len(num) >= 2 else numbers_in_words[num[0]]

def check_and(num):
    if int(num[-2:]) > 0:
        return and_word + " "
    return ""

user_input = input("Please enter a number from 0 to 99999:")

if 0 <= int(user_input) <= 99999:
    user_input_words = convert_thousands(user_input) + \
                       convert_hundreds(user_input) + \
                       check_and(user_input) + \
                       convert_tens_ones(user_input)
    
    result = user_input_words[0].upper() + user_input_words[1:].lower()
    
    print(f'{user_input = }')
    print(result.strip())
else:
    print("Please enter a number between 0 and 99999.")
```
