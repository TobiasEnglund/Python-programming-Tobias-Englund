# Matplotlib

```python
import matplotlib.pyplot as plt

x = list(range(-10, 11))
y = [value**2 for value in x]
y2 = [value**3 for value in x]

print(f'{x = }')
print(f'{y = }')

plt.plot(x, y, label = 'y = x^2', color = "blue")
plt.plot(x, y2, label = 'y = x^3', color = "orange")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2 and y = x^3')
plt.grid(True)
plt.legend()
```

# Bars

```python
labels = ["Male", "Female"]
values = [15, 23]

plt.bar(labels, values)
```

## Pie charts

```python
plt.pie(values, labels = labels)
```

## Sub plots

```python
x = list(range(-10, 11))

y = [value**2 for value in x]
y2 = [value**3 for value in x]

fig, axs = plt.subplots(1, 2)

# Första grafen
axs[0].plot(x, y, color='blue')
axs[0].set_facecolor('lightgray')

# Andra grafen
axs[1].plot(x, y2, color='red')
axs[1].set_facecolor('lightgray')

# Bakgrund för hela figuren
fig.set_facecolor("white")
```
