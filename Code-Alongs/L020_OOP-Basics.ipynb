# Object Oriented Programming (OOP)
- A "class" is a blue print that can be used to create objects
- An "instance" is the actual object containing data and methods

```python
student_a = {
    "FirstName": "Fredrik",
    "LastName": "Johansson",
    "Age": 42,
    "Grade": 3.4}
student_b = {
    "FirstName": "Anna",
    "LastName": "Karlsson",
    "Age": 24,
    "Grade": 4.2}

students = [student_a, student_b]

def set_grade(student, grade):
    student["Grade"] = grade
    
def get_name(student):
    return f'{student["FirstName"]} {student["LastName"]}'

for student in students:
    set_grade(student, 3.0)
    print(get_name(student))
```

```python
class Student:
    def __init__(self, name, grade):
        self.grade = grade
        self.name = name
        
    def set_pet(self, name, age):
        self.pet = Cat(name, age)
        
    def set_grade(self, grade):
        self.grade = grade
        
```

```python

student_a = Student("Fredrik", 4.5)
student_b = Student("Anna", 3.3)

print(student_a.__dict__)
print(student_b.__dict__)
print()

Student.set_grade(student_a, 4.6)
student_b.set_grade(3.2)
student_a.age = 42

print(student_a.__dict__)
print(student_b.__dict__)
```

```python
class Cat:
    def __init__(self, name, age, best_friend = None):
        self.name = name
        self.age = age
        self.best_friend = best_friend
        
    def greet(self, cat_friend = None): 
        if cat_friend == None:
            print(f'Hello, I \'m {self.name}, and I\'m {self.age} years old.')
        else:
            age_difference = abs(cat_friend.age - self.age)
            year_string = "year" if age_difference == 1 else "years"
            
            if cat_friend.age < self.age:
                print(f'Hi {cat_friend.name}, my name is {self.name} and I\'m {age_difference} {year_string} older than you!')
            elif cat_friend.age == self.age:
                print(f'Hi {cat_friend.name}, my name is {self.name} and I\'m as old as you!')
            elif cat_friend.age > self.age:
                print(f'Hi {cat_friend.name}, my name is {self.name} and I\'m {age_difference} {year_string} younger than you!')
                
    def print_best_friend(self):
        if self.best_friend is None:
            print(f'My name is {self.name} and I don\'t have a best friend :(')
        else:
            print(f'My name is {self.name} and my best friend is {self.best_friend.name} :)')
            
    # def print_best_friend(self):
    #     if not hasattr(self, "best_friend"): return
    #     if self.best_friend is None: return
        
    #     print(f'My name is {self.name} and my best friend is {self.best_friend.name} :)')
```

```python
cats = []
cats.append(Cat("Misse", 3))
cats.append(Cat("Kisse", 2))
cats.append(Cat("Mjau", 5))
cats.append(Cat("Jens", 5))

cats[0].best_friend = cats[2]
cats[1].best_friend = cats[1]
cats[2].best_friend = cats[0]
cats[3].best_friend = None

for cat in cats:
    cat.print_best_friend()

print()
for cat in cats:
    cat.greet()
    
print()
for cat in cats[1:]:
    cats[0].greet(cat)
    
# my_cat.greet() 
# my_cat.greet(my_other_cat)
# my_cat.greet(my_third_cat)
# my_third_cat.greet(my_fourth_cat)
```
